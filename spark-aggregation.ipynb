```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import datediff, to_date, unix_timestamp
```

```python
spark = SparkSession.builder.appName('Practice').getOrCreate()
```

```python
df_spark = spark.read.parquet('C:\\Users\\cihat\\Downloads\\fhvhv_tripdata_2023-01.parquet', header=True)
```

```python
df_spark.count()
```

```python
df_spark.show(5)
```

```python
same_location = df_spark.filter(df_spark['PULocationID'] == df_spark['DOLocationID'])
```

```python
same_location.show(5)
```

```python
date_diff = same_location.withColumn("minute_diff", (unix_timestamp(same_location["dropoff_datetime"]) - unix_timestamp(same_location["pickup_datetime"])) / 60)
```

```python
date_diff.orderBy("minute_diff", ascending=False).show(5)
```

```python
date_diff.select(['PULocationID', 'DOLocationID', 'minute_diff']).orderBy("minute_diff", ascending=False).show(10)
```

```python
date_diff.select(["trip_miles", "minute_diff"]).orderBy("minute_diff", ascending=False).show(10)
```
